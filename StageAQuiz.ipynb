```python
import pandas as pd
import numpy as np 
```

```python
# Getting the encoding of the csv 
with open('FoodBalanceSheets_E_Africa_NOFLAG.csv') as f:
    print(f)
    
# Reading the csv 
data = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding = 'cp1252')
```

```python
#Q11
# What is the total sum of Animal Fat produced in 2014 and 2017 respectively 
# Ans: 209460.54 & 269617.53

# performing aggregation using aggregate method
 
grp1 = data.groupby('Item')
 
grp1.aggregate(np.sum)
```

```python
# Q12 
# What is the mean and standard deviation across the whole dataset
# Ans: 135.236 & 1603.404


data.describe()
```

```python
# Q13 
# What is teh total number and percentage of missing data in 2016 to 2 decimal places 
# Ans: 1535 & 2.58%

missing_data = data['Y2016'].isnull().sum()

percentage = (( missing_data / data['Y2016'].count())* 100)

print(missing_data)
print(percentage)

```
